```python
from requests import get
from bs4 import BeautifulSoup
```

```python
base_url = "https://quotes.toscrape.com/page/1/"

response = get(base_url)
```

```python
soup = BeautifulSoup(
    response.content,
    "html.parser"
)
```

```python
quotes = soup.find_all("div", attrs={"class": "quote"})
```

```python
for index, quote in enumerate(quotes):
    print(quote.text)

    with open(f"quotes/quotes_{index}“.txt", "a") as f:
        f.write(quote.text)
        f.write("\n\n")
```
